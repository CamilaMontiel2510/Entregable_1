## Series
### Una serie es una matriz **etiquetada** unidimensional capaz de contener cualquier tipo de dato( enteros, cadenas, flotantes, objetos Python, etc. ). Pero a diferencia de un vector usual de numpy, este cuenta con una columna o eje de indexación.

### Se invoca a traves del metodo _pd.Series(data, index = indices)_ donde data puede ser difernetes dtipos de elementos:
   * Diccionarios de Pyhton
   * Listas de Python
   * Un Valor escalar
   * Un arreglo de numpy (ndarray)


```python
np.random.randn(5)
```

```python
d = {"b": 1, "a": 0, "c": 2}
ds1 = pd.Series(d)
print(ds1)
print("\n")
print(ds1.index,"\n")
ds2 = pd.Series(np.random.randn(5), index=np.arange(5,15,2))
print(ds2,"\n")
ds3 = pd.Series("hola", index=["a", "b", "c", "d", "e"])
print(ds3,"\n")
ds4 = pd.Series([5,3,3,4,6], index=["a", "b", "c", "d", "e"])
print(ds4)
```

## Los nombres usados son los siguientes
* 'i1': Entero con signo de 8 bits
* 'i2': Entero con signo de 16 bits
* 'i4': Entero con signo de 32 bits
* 'i8': Entero con signo de 64 bits
* 'u1': Entero sin signo de 8 bits
* 'u2': Entero sin signo de 16 bits
* 'u4': Entero sin signo de 32 bits
* 'u8': Entero sin signo de 64 bits
* 'f2': Punto flotante de 16 bits
* 'f4': Punto flotante de 32 bits
* 'f8': Punto flotante de 64 bits
* bool_: Tipo booleano.
* object_: Tipo de objeto de Python, que permite almacenar cualquier objeto de Python en un array NumPy.
* datetime64: Tipo de datos que representa fechas y horas en unidades de tiempo de 64 bits.
* timedelta64: Tipo de datos que representa una duración o diferencia de tiempo entre dos datetime64.
* void: Tipo de datos estructurados sin formato que pueden contener varios tipos de datos.
